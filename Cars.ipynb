Importing Libraries

```python
import pandas as pd
import numpy
```

Loading the dataset

```python
i='https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars1.csv'
j='https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars2.csv'
```

Calling cars1 and cars2 and reading data

```python
cars1=pd.read_csv(i)
cars2=pd.read_csv(j)
cars2.head()
```

Printing the top rows of cars1

```python
cars1.head()
```

Removing the unnamed columns

```python
cars1.drop(cars1.columns[cars1.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)
cars1.head()
```

Checking the total number of rows

```python
cars1.shape[0],cars2.shape[0]
```

Combining data

```python
x=[cars1,cars2]
cars=pd.concat(x)
cars.shape[0]
```

Adding new column to data cars

```python
owner = pd.DataFrame(numpy.random.randint(15000,73000,size=398))
cars['owners']=owner
cars.head()
```
