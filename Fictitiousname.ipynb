Importing the Libraries

```python
import pandas as pd
import numpy as np
```

Creating the dataframes

```python
raw_data_1 = {
        'subject_id': ['1', '2', '3', '4', '5'],
        'first_name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'], 
        'last_name': ['Anderson', 'Ackerman', 'Ali', 'Aoni', 'Atiches']}

raw_data_2 = {
        'subject_id': ['4', '5', '6', '7', '8'],
        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'], 
        'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}

raw_data_3 = {
        'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'test_id': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]}
```

Loading Data

```python
data1=pd.DataFrame(raw_data_1)
data2=pd.DataFrame(raw_data_2)
data3=pd.DataFrame(raw_data_3)
```

Combining Data along rows

```python
all_data = pd.concat([data1,data2])
all_data.head()
```

Combining Data along columns

```python
all_data_col = pd.concat([data1,data2],axis=1)
all_data_col.head()
```

Printing Data3

```python
data3.head()
```

Combining Data on subject id

```python
x=pd.merge(all_data,data3,on='subject_id')
x.head(10)
```

merging data1 and data2 on subject id

```python
y=pd.merge(data1,data2,on='subject_id',how='inner')
y.head()
```

Merge all values in data1 and data2, with matching records from both sides where available.

```python
z=pd.merge(data1,data2,on=list(data1.columns),how='inner')
z.head()
```
