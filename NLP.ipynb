```python
import spacy
import textacy
```

STEP 1: Importing or Loading the text

```python
dic = spacy.load('en_core_web_lg')
f = open('file.txt','rt')
```

STEP 2: Sentence Segmentation

```python
data = f.read()
doc = dic(data)
for s in doc.sents:
    print(s)
```

STEP 3: Tokenization

```python
for x in doc:
  print(x.text)
```

STEP 4: Parts of Speech Tagging

```python
for token in doc:
  print(token, token.pos_)
```

STEP 5: Lemmatization

```python
for token in doc:
  print(token, token.lemma_)
```

STEP 6: Stop Words

```python
stopwords = dic.Defaults.stop_words
lst=[]
for token in data.split():
    if token.lower() not in stopwords:
        lst.append(token)
```

```python
print(lst)
```

STEP 7: Dependency Prasing

```python
for token in doc:
  print(token.text, token.tag_, token.head.text, token.dep_)
```

STEP 8: Noun Phrases

```python
for word in doc.noun_chunks:
  print(word.text)
```

STEP 9: Named Entity Recognition

```python
for word in doc.ents:
  print(word.text,word.label_)
#displacy.render(doc, style="ent")
```

STEP 10: Coreference Resolution

```python
statements = textacy.extract.semistructured_statements(doc,entity="India", cue='be') 
print('This is all we know about India')
for statement in statements:
  subject, verb, fact = statement 
  print(f"-  {fact}")
```
