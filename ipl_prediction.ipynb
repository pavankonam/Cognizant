Loading the complete ball to ball dataset of IPL match from 2008 to 2020.

```python
import pandas as pd
import numpy as np
```

Planning to estimate the run rate of a player vs the team separately in first innings and seconds innings and calculating their average for an over

```python
data = pd.read_csv('IPL Ball-by-Ball 2008-2020.csv')
data.head()
```

Taking the data of Chennai Super Kings batting played against Mumbai Indians

```python
cs = data.loc[data['batting_team']=='Chennai Super Kings']
cs = cs.loc[cs['bowling_team']=='Mumbai Indians']
cs.head()
```

Removing the unwanted columns
I am estimating the overall player performance against the team this can be improved by considering the bowlera and non striker end statistics

```python
cs= cs.drop(['id','non_striker','is_wicket','dismissal_kind','player_dismissed','fielder','extras_type'],axis=1)
cs.head()
```

Checking all the batsman played in csk

```python
cs.batsman.unique()
```

Separating the data as innings 1 and 2 as the player mindset and game methodology change as per the innings

```python
a= cs.loc[cs['inning']== 1]
b= cs.loc[cs['inning']== 2]
a.head()
```

separating the data of MS Dhoni played in innings1 and 2

```python
dhonia = a.loc[a['batsman']=='MS Dhoni']
dhonib = b.loc[b['batsman']=='MS Dhoni']
dhonib.shape
```

Dividing the total runs scored by MS Dhoni to total no of balls played to get the average score of Dhoni in an over or ball to ball.

```python
dh1=dhonia['total_runs'].sum()/len(dhonia)
dh2 = dhonib['total_runs'].sum()/len(dhonib)
```

Similarly creating the data of all the players in CSK

```python
z =cs.batsman.unique()
df = pd.DataFrame(z,columns=['players'])
```

```python
innings1 = []
innings2 = []
for i in range(len(z)):
  x= a.loc[a['batsman']==z[i]]
  y= b.loc[b['batsman']==z[i]]
  l=len(x)
  m=len(y)
  innings1.append(((x['batsman_runs'].sum()/l)*6) if l!=0 else 0)
  innings2.append(((y['batsman_runs'].sum()/m)*6) if m!=0 else 0)
print(innings1)
```

```python
df['innings1'] = np.array(innings1).tolist()
df['innings2'] = np.array(innings2).tolist()
print(df)
```
